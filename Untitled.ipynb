```python
df_train=pd.read_csv('train.csv')
```

```python
df_dogs_train=df_train[df_train['AnimalType']=='Dog']
df_dogs_train.drop('AnimalID', inplace=True, axis=1)
df_dogs_train.drop('Name', inplace=True, axis=1)
df_dogs_train.drop('OutcomeSubtype', inplace=True, axis=1)
df_dogs_train.drop('DateTime', inplace=True, axis=1)
df_dogs_train.drop('AnimalType', inplace=True, axis=1)
df_dogs_train.columns

df_dogs_train.dropna(inplace=True)
```

```python
'''
outcomes = {}
for _, row in df_dogs_train.iterrows():
    if row['OutcomeType'] in outcomes:
        row['OutcomeType'] = outcomes[row['OutcomeType']]
    else:
        print(len(outcomes))
        #print(f'{outcomes[row["OutcomeType"]]} = {len(outcomes+1)}')
        outcomes[row['OutcomeType']] = len(outcomes)+1
        row['OutcomeType'] = outcomes[row['OutcomeType']]
        print(len(outcomes))
'''
'''
animaltypes = {}
for _, row in df_dogs.iterrows():
    if row['AnimalType'] in animaltypes:
        row['AnimalType'] = animaltypes[row['AnimalType']]
    else:
        print(len(animaltypes))
        animaltypes[row['AnimalType']] = len(animaltypes)+1
        row['AnimalType'] = animaltypes[row['AnimalType']]
        print(len(animaltypes))'''
        
'''
SexuponOutcome = {}
for _, row in df_dogs.iterrows():
    if row['SexuponOutcome'] in SexuponOutcome:
        row['SexuponOutcome'] = SexuponOutcome[row['SexuponOutcome']]
    else:
        SexuponOutcome[row['SexuponOutcome']] = len(SexuponOutcome)+1
        row['SexuponOutcome'] = SexuponOutcome[row['SexuponOutcome']]'''

SexuponOutcome = {}
# df_dogs_train.drop(df_dogs_train[df_dogs_train['SexuponOutcome'] == 'Unknown'], inplace=True)
for _, row in df_dogs_train.iterrows():
    if row['SexuponOutcome'] in SexuponOutcome:
        row['SexuponOutcome'] = SexuponOutcome[row['SexuponOutcome']]
    else:
        SexuponOutcome[row['SexuponOutcome']] = len(SexuponOutcome)+1
        row['SexuponOutcome'] = SexuponOutcome[row['SexuponOutcome']]


# AgeuponOutcome fix to month

AgeuponOutcome = {}
for _, row in df_dogs_train.iterrows():
    if type(row['AgeuponOutcome']) == float:
        row['AgeuponOutcome'] = 0
        if row['AgeuponOutcome'] not in AgeuponOutcome:
            AgeuponOutcome[row['AgeuponOutcome']] = 0
    
    elif row['AgeuponOutcome'].split()[1] == 'month' or row['AgeuponOutcome'].split()[1] == 'months':
        row['AgeuponOutcome'] = 0
        if row['AgeuponOutcome'] not in AgeuponOutcome:
            AgeuponOutcome[row['AgeuponOutcome']] = 0
            
    else:
        row['AgeuponOutcome'] = row['AgeuponOutcome'].split()[0]
        if row['AgeuponOutcome'] not in AgeuponOutcome:
            AgeuponOutcome[row['AgeuponOutcome']] = row['AgeuponOutcome'].split()[0]
        


Breed = {}
for _, row in df_dogs_train.iterrows():
    if row['Breed'] in Breed:
        row['Breed'] = Breed[row['Breed']]
    else:
        Breed[row['Breed']] = len(Breed)+1
        row['Breed'] = Breed[row['Breed']]

Color = {}
for _, row in df_dogs_train.iterrows():
    if row['Color'] in Color:
        row['Color'] = Color[row['Color']]
    else:
        Color[row['Color']] = len(Color)+1
        row['Color'] = Color[row['Color']]
        
```
